```python
food_value = img_value[img_value["is_nonfood"] == 0]
food_value = food_value.dropna(subset=["id"])
food_value[food_value == "data not acquired"] = np.nan
food_value[["id", "is_nonfood","is_processed_foods"]] =food_value[["id", "is_nonfood","is_processed_foods"]].astype(int)
food_value[["Item_description"]] = food_value[["Item_description"]].astype(str)
float_columns = [col for col in dtype_dict.keys() if dtype_dict[col] == 'float']
food_value[float_columns] = food_value[float_columns].astype(float)
food_value.info()
```

```python
type_description = "type:1=fruit, 2=with vegetables, 3=chocolate, 4=with fish, 5=with meat, 6=nuts, 7=drinks, 21=Flowers/leaves, 22=Animals, 23=Tools,  24=household (non-kitchen), 25=Kitchen Utensils, 26=office, 27=food packaging, 28=other"
type_description = type_description.replace("type:", "").replace("\n", "").split(", ")
type_dict = {int(t.split("=")[0]): t.split("=")[1] for t in type_description if "=" in t}
type_dict
```

```python
food_value["type"] = food_value["type"].map(type_dict)
food_value = pd.get_dummies(food_value, columns=["type"], prefix="is_type" ,dtype=int)
```
